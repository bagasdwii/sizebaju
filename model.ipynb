```python
mapping_size = {
    'XXS' : 1,
    'S' : 2,
    'M' : 3,
    'L' : 4,
    'XL' : 5,
    'XXL' : 6,
    'XXXL' : 7
}

df['size'] = df['size'].map(mapping_size)
df
```

```python
df.isna().sum()
```

```python
# Mengisi data yang  hilang menggunakan median 
df["age"] = df["age"].fillna(df['age'].median())
df["height"] = df["height"].fillna(df['height'].median())
df["weight"] = df["weight"].fillna(df['weight'].median())
df.isna().sum()
```

```python
df["bmi"] = df["weight"]/(df["height"]**2)
df.head()
```

```python

df = pd.read_csv('final_test_10k.csv')
# Pisahkan fitur (X) dan target (y)
X =df.drop('size', axis=1)  # Fitur
y = df['size']  # Target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
# Misalkan data Anda ada dalam variabel X_train dan y_train
# Contoh pembuatan model Decision Tree
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
# Simpan model ke dalam file .sav
filename = 'DecisionTree_Classifier.sav'
pickle.dump(model, open(filename, 'wb'))
# Load model dari file .sav
loaded_model = pickle.load(open('DecisionTree_Classifier.sav', 'rb'))
```

```python
y_pred = loaded_model.predict(X_test)

# Hitung akurasi menggunakan nilai sebenarnya (y_test) dan hasil prediksi (y_pred)
accuracy = accuracy_score(y_test, y_pred)
print(f"Akurasi model Decision Tree: {accuracy:.2f}")
```
